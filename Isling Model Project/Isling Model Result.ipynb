```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# First Example Run

```python
# Parameters of run
# L = 50
# n_warmup = 70
# n_step = 2
# n_meas = 50
# congifured_n_bins = 10

# Magnitization and energy of the lattice over time

with open("mag_vector_1.csv","r") as fh:
    mag_vector = fh.readlines()
    
with open("energy_vector_1.csv","r") as fh:
    energy_vector = fh.readlines()
    
    
for c in range(len(mag_vector)):
    mag_vector[c] = mag_vector[c].replace("\n","")
    mag_vector[c] = int(mag_vector[c])
    
    energy_vector[c] = energy_vector[c].replace("\n","")
    energy_vector[c] = int(energy_vector[c])

fig, ax1 = plt.subplots(figsize=(12,7))

color = 'tab:blue'
ax1.set_xlabel('Time (Lattice Measurements)')
ax1.set_ylabel('Magnetization', color=color)
ax1.plot(mag_vector,color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

color = 'tab:orange'
ax2.set_xlabel('Time (Lattice Measurements)')
ax2.set_ylabel('Energy', color=color)
ax2.plot(energy_vector,color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title("Lattice Energy and Magnetization Through Time")
plt.show()
```

```python
plt.figure(figsize=(12,7))

plt.hist(mag_vector,bins=int(round(len(mag_vector)**(1/2))),label="Magnetization")
plt.hist(energy_vector,bins=int(round(len(energy_vector)**(1/2))),label="Energy")
plt.legend()
plt.title("Lattice Energy and Magnetization Distribution")
plt.show()
```

```python
# Magnetization and energy averages over time

df = pd.read_csv("lattice_1.csv")
df = df.dropna()

magnetization_average = list(df["M"])
energy_average = list(df["E"])

magnetization_error = list(df["M_err"])
energy_error = list(df["E_err"])

fig, ax1 = plt.subplots(figsize=(12,7))

t_range = [t for t in range(len(energy_average))]
 
color = 'tab:blue'
ax1.set_xlabel('Time (Lattice Measurements)')
ax1.set_ylabel('Magnetization Average', color=color)
ax1.errorbar(t_range,magnetization_average,magnetization_error, marker='s',color=color,
             markeredgecolor="blue",markerfacecolor="blue",ms=2)
#ax1.scatter(t_range,magnetization_average,color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

color = 'tab:orange'
ax2.set_xlabel('Time (Lattice Measurements)')
ax2.set_ylabel('Magnetization Average', color=color)
ax2.errorbar(t_range,energy_average,energy_error, marker='s',color=color,
             markeredgecolor="orange",markerfacecolor="orange",ms=2)
#ax2.scatter(t_range,energy_average,color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title("Lattice Energy and Magnetization Through Time")
plt.show()
```

```python
plt.figure(figsize=(12,7))

plt.hist(magnetization_average,bins=int(round(len(magnetization_average)**(1/2))),label="Magnetization Average")
plt.hist(energy_average,bins=int(round(len(energy_average)**(1/2))),label="Energy Average")
plt.legend()
plt.title("Lattice Energy and Magnetization Distribution")
plt.show()
```
